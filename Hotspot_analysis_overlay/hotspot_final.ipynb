```python
SALES_FILE = "suwon_hotspots_gi_scores.csv"  # 네 파일 경로

# 기준 (필요하면 숫자만 조정)
Z_HOT  = 1.65     # ~90%
Z_COLD = -1.65
P_MAX  = 0.10     # p <= 0.10일 때만 유의로 볼지 여부(있을 때 적용)

df = pd.read_csv(SALES_FILE)

def pick_hot(df, z_col, p_col=None, label_col=None):
    if label_col and label_col in df.columns:
        s = df[label_col].astype(str).str.lower()
        cond = s.str.contains("hot", na=False)
    else:
        cond = df[z_col] >= Z_HOT
        if p_col and p_col in df.columns:
            cond &= (df[p_col] <= P_MAX)
    return df.loc[cond].copy()

def pick_cold(df, z_col, p_col=None, label_col=None):
    if label_col and label_col in df.columns:
        s = df[label_col].astype(str).str.lower()
        cond = s.str.contains("cold", na=False)
    else:
        cond = df[z_col] <= Z_COLD
        if p_col and p_col in df.columns:
            cond &= (df[p_col] <= P_MAX)
    return df.loc[cond].copy()

def radius_from_z(z):
    z = max(float(z), 0.0)
    return float(min(36, max(6, 6 + 1.8*z)))  # 카카오 원 반경 스케일값(상대)

def export_sales_layer(df_sub, z_col, color_hex, out_path):
    out = df_sub[["lat","lon", z_col]].copy()
    out.rename(columns={z_col:"z"}, inplace=True)
    out["color"]  = color_hex
    out["radius"] = [radius_from_z(v) for v in out["z"]]
    out[["lat","lon","color","radius","z"]].to_json(out_path, orient="records", force_ascii=False)
    print("저장:", out_path, len(out))

# AMT 기준
amt_hot  = pick_hot(df,  "amt_z", "amt_p", "amt_label")
amt_cold = pick_cold(df, "amt_z", "amt_p", "amt_label")
export_sales_layer(amt_hot,  "amt_z", "#FB8C00", "sales_amt_hot.json")   # 주황
export_sales_layer(amt_cold, "amt_z", "#1976D2", "sales_amt_cold.json")  # 파랑

# NOC 기준
noc_hot  = pick_hot(df,  "noc_z", "noc_p", "noc_label")
noc_cold = pick_cold(df, "noc_z", "noc_p", "noc_label")
export_sales_layer(noc_hot,  "noc_z", "#8E24AA", "sales_noc_hot.json")   # 보라
export_sales_layer(noc_cold, "noc_z", "#039BE5", "sales_noc_cold.json")  # 하늘
```

```python
SALES_FILE = "suwon_hotspots_gi_scores.csv"  # 네 파일 경로로 둠

# ---- 기준 (원하면 숫자만 조정) ----
Z_HOT  = 1.65      # ~90%
Z_COLD = -1.65
P_MAX  = 0.10      # p <= 0.10일 때 유의(해당 컬럼 있을 때 적용)

df = pd.read_csv(SALES_FILE)

def pick_hot(df, z_col, p_col=None, label_col=None):
    if label_col and label_col in df.columns:
        s = df[label_col].astype(str).str.lower()
        cond = s.str.contains("hot", na=False)
    else:
        cond = df[z_col] >= Z_HOT
        if p_col and p_col in df.columns:
            cond &= (df[p_col] <= P_MAX)
    return df.loc[cond].copy()

def pick_cold(df, z_col, p_col=None, label_col=None):
    if label_col and label_col in df.columns:
        s = df[label_col].astype(str).str.lower()
        cond = s.str.contains("cold", na=False)
    else:
        cond = df[z_col] <= Z_COLD
        if p_col and p_col in df.columns:
            cond &= (df[p_col] <= P_MAX)
    return df.loc[cond].copy()

# ---------- z → color / radius ----------
def color_from_z(z: float) -> str:
    z = float(z)
    if z >=  2.58: return "#B71C1C"  # Hot 99
    if z >=  1.96: return "#D32F2F"  # Hot 95
    if z >=  1.65: return "#FB8C00"  # Hot 90
    if z >   0.00: return "#FFE0B2"  # 약한 양성
    if z <= -2.58: return "#0D47A1"  # Cold 99
    if z <= -1.96: return "#1976D2"  # Cold 95
    if z <= -1.65: return "#42A5F5"  # Cold 90
    return "#BBDEFB"                 # 약한 음성

def radius_from_z(z: float) -> float:
    # |z|가 클수록 크게 (6~40)
    return float(np.clip(8 + 4*abs(float(z)), 6, 40))

def export_layer(df_sub, z_col, out_path):
    out = df_sub[["lat","lon", z_col]].copy()
    out.rename(columns={z_col:"z"}, inplace=True)
    out["color"]  = out["z"].map(color_from_z)
    out["radius"] = out["z"].map(radius_from_z)
    out[["lat","lon","color","radius","z"]].to_json(out_path, orient="records", force_ascii=False)
    print("저장:", out_path, len(out))

# ---------- AMT ----------
amt_hot  = pick_hot(df,  "amt_z", "amt_p", "amt_label")
amt_cold = pick_cold(df, "amt_z", "amt_p", "amt_label")
export_layer(amt_hot,  "amt_z", "sales_amt_hot.json")
export_layer(amt_cold, "amt_z", "sales_amt_cold.json")

# ---------- NOC ----------
noc_hot  = pick_hot(df,  "noc_z", "noc_p", "noc_label")
noc_cold = pick_cold(df, "noc_z", "noc_p", "noc_label")
export_layer(noc_hot,  "noc_z", "sales_noc_hot.json")
export_layer(noc_cold, "noc_z", "sales_noc_cold.json")

print("완료: sales_amt_hot.json, sales_amt_cold.json, sales_noc_hot.json, sales_noc_cold.json")
```
